```python
import asyncio
import websockets
import subprocess

# 你的 Lichess token，請替換成你自己的
TOKEN = "在這裡填入你的 Token"

# Stockfish 的執行檔路徑（如果有加到環境變數，可以直接寫 "stockfish"）
STOCKFISH_PATH = "D:/lichess2/UndergraduationProject/stockfish/stockfish-windows-x86-64-avx2.exe"

async def run_stockfish():
    # 啟動 Stockfish 引擎
    process = await asyncio.create_subprocess_exec(
        STOCKFISH_PATH,
        stdin=asyncio.subprocess.PIPE,
        stdout=asyncio.subprocess.PIPE
    )
    return process

async def main():
    uri = "wss://lichess.org/api/bot/ws/engine"
    headers = {"Authorization": f"Bearer {TOKEN}"}

    # 建立與 Lichess 的 WebSocket 連線
    async with websockets.connect(uri, extra_headers=headers) as websocket:
        engine = await run_stockfish()

        async def read_from_engine():
            while True:
                line = await engine.stdout.readline()
                if not line:
                    break
                decoded_line = line.decode().strip()
                print(f"[Stockfish] {decoded_line}")
                await websocket.send(decoded_line)

        # 啟動一個協程，持續從 Stockfish 讀取輸出
        asyncio.create_task(read_from_engine())

        # 從 Lichess 持續接收指令，並轉發給 Stockfish
        async for message in websocket:
            print(f"[Lichess] {message}")
            engine.stdin.write((message + "\n").encode())
            await engine.stdin.drain()

asyncio.run(main())
```

```python
import requests

# 要查詢的位置（FEN 格式）
fen = "r1bqkbnr/pppp1ppp/2n5/1B2p3/4P3/5N2/PPPP1PPP/RNBQK2R b KQkq - 3 3"

# 設定 API 網址與參數
url = "https://lichess.org/api/cloud-eval"
params = {
    "fen": fen,
    "multiPv": 1,     # 要查幾個最佳著法（最多 5 個）
    "syzygy": "true"  # 是否使用 Syzygy tablebase（用於結尾局面）
}

# 發送 GET 請求
response = requests.get(url, params=params)

# 檢查狀態並顯示結果
if response.status_code == 200:
    data = response.json()
    print("最佳著法：", data["pvs"][0]["moves"])
    print("評估值（以 centipawn 為單位）：", data["pvs"][0]["cp"])
    print("深度：", data["depth"])
else:
    print("查詢失敗，狀態碼：", response.status_code)
```

```python
import pandas as pd
import chess
import requests
import re
import os
import time

input_folder = "database/Filter_CSV/180+2"
output_folder = "database/Filter_evaluate_CSV/180+2"

os.makedirs(output_folder, exist_ok=True)

file_path = os.path.join(input_folder, "chunk_1.csv")
df = pd.read_csv(file_path)

# 取出第一筆棋局的走法
moves_raw = df.loc[1, 'MovesProcessed']

# 移除 "1...", "2..." 等對手回合標示
moves_clean = re.sub(r'\d+\.\.\.', '', moves_raw)
# 移除 "1. e4" 中的 move number，只留下 e4
moves_only = re.sub(r'\d+\.\s*', '', moves_clean).strip()
# 分割成單一步驟走法
moves = moves_only.split()

print(moves)
# 初始化棋盤並生成每一步後的 FEN
board = chess.Board()
fens = []
for move in moves:
    try:
        board.push_san(move)
        fens.append(board.fen())
    except:
        break

# 使用 Lichess Cloud Evaluation API 查詢每個 FEN 的評估值
def get_eval(fen):
    url = 'https://lichess.org/api/cloud-eval'
    params = {'fen': fen, 'multiPv': 1}
    try:
        res = requests.get(url, params=params)
        if res.status_code == 200:
            data = res.json()
            cp = data['pvs'][0].get('cp')  # 若是找不到會回傳 None
            mate = data['pvs'][0].get('mate')
            return f"M{mate}" if mate else cp
        else:
            return None
    except:
        return None


# 取得每一步的評估值（會花時間，注意 API 限制）
evals = [get_eval(fen) for fen in fens]

# 加入新欄位到 DataFrame
df.loc[1, 'EvalByMove'] = str(evals)

# 儲存成新的 CSV
df.to_csv('chunk_1_with_eval.csv', index=False)
print("已儲存成 'chunk_1_with_eval.csv'")
```

```python
import os
import pandas as pd
import re
import time
from stockfish import Stockfish
from tqdm import tqdm

# ========== 設定 ==========
stockfish_path = "D:/lichess2/UndergraduationProject/stockfish/stockfish-windows-x86-64-avx2.exe"

input_folder = "database/Filter_CSV/180+2"
output_folder = "database/Filter_evaluate_CSV/180+2"

os.makedirs(output_folder, exist_ok=True)

# stockfish初始化
def create_stockfish(depth):
    engine = Stockfish(path=stockfish_path, depth=depth)
    engine.update_engine_parameters({
        "Threads": 2,
        "Minimum Thinking Time": 30
    })
    return engine

# 處理Moves欄位：清除move number
def convert_moves_for_stockfish(moves_str):
    cleaned_moves = re.sub(r'\d+\.\.\.|(\d+\.)', '', moves_str)
    moves = cleaned_moves.split()
    return moves

# ========== 開始處理 ==========

# 讀取 chunk_1.csv
file_path = os.path.join(input_folder, "chunk_1.csv")
df = pd.read_csv(file_path)

# 只取第0行
moves_str = df.loc[0, "MovesProcessed"]
moves_list = convert_moves_for_stockfish(moves_str)
print(moves_list)

# 準備stockfish引擎
stockfish_5 = create_stockfish(5)
stockfish_15 = create_stockfish(15)

# 儲存評估值
evals_depth_5 = []
evals_depth_15 = []

start_time = time.time()

# 深度5
print("🔵 深度5開始計算...")
for i in tqdm(range(1, len(moves_list)+1), desc="Depth 5 Progress"):
    stockfish_5.set_position(moves_list[:i])
    eval_info = stockfish_5.get_evaluation()
    if eval_info["type"] == "cp":
        evals_depth_5.append(eval_info["value"])
    elif eval_info["type"] == "mate":
        evals_depth_5.append(10000 if eval_info["value"] > 0 else -10000)
    else:
        evals_depth_5.append(0)  # 出現未知型別，安全地補0

# 深度15
print("\n🟠 深度15開始計算...")
for i in tqdm(range(1, len(moves_list)+1), desc="Depth 15 Progress"):
    stockfish_15.set_position(moves_list[:i])
    eval_info = stockfish_15.get_evaluation()
    if eval_info["type"] == "cp":
        evals_depth_15.append(eval_info["value"])
    elif eval_info["type"] == "mate":
        evals_depth_15.append(10000 if eval_info["value"] > 0 else -10000)
    else:
        evals_depth_15.append(0)

end_time = time.time()

# 將結果寫進df
df.loc[0, "Evaluation_Depth_5"] = str(evals_depth_5)
df.loc[0, "Evaluation_Depth_15"] = str(evals_depth_15)

# 儲存到新的csv
output_path = os.path.join(output_folder, "chunk_1_test.csv")
df.to_csv(output_path, index=False)

print(f"\n 測試結束，檔案儲存於：{output_path}")
print(f" 總耗時：{end_time - start_time:.2f} 秒")
```

```python
import os
import re
import time
import pandas as pd
import chess
from stockfish import Stockfish
from tqdm import tqdm

# 初始化兩個 Stockfish 引擎，分別用不同深度
stockfish_depth_5 = Stockfish(path="D:/lichess2/UndergraduationProject/stockfish/stockfish-windows-x86-64-avx2.exe", depth=5)
stockfish_depth_5.update_engine_parameters({
    "Threads": 2,
    "Minimum Thinking Time": 30
})

stockfish_depth_15 = Stockfish(path="D:/lichess2/UndergraduationProject/stockfish/stockfish-windows-x86-64-avx2.exe", depth=15)
stockfish_depth_15.update_engine_parameters({
    "Threads": 2,
    "Minimum Thinking Time": 30
})

# 處理 Moves 欄位：清除 move number
def convert_moves_for_stockfish(moves_str):
    cleaned_moves = re.sub(r'\d+\.\.\.|(\d+\.)', '', moves_str)
    moves = cleaned_moves.split()
    return moves

# 讀取 chunk_1.csv
input_folder = "database/Filter_CSV/180+2"  # ⚡記得換
file_path = os.path.join(input_folder, "chunk_1.csv")
df = pd.read_csv(file_path)

# 新增兩個欄位
df["Evaluation_Depth_5"] = None
df["Evaluation_Depth_15"] = None

# 開始計時
start_time = time.time()

# 用 tqdm 顯示進度條
for idx in tqdm(df.index, desc="Processing Games"):
    moves_str = df.loc[idx, "MovesProcessed"]
    # 如果是 NaN 或空字串，跳過
    if pd.isna(moves_str) or moves_str.strip() == "":
        print(f"Game {idx} skipped (no moves)")
        continue

    #if(idx>=5) : break
        
    moves_list = convert_moves_for_stockfish(moves_str)

    board = chess.Board()
    evaluations_depth_5 = []
    evaluations_depth_15 = []

    for move in moves_list:
        # 設定盤面
        stockfish_depth_5.set_fen_position(board.fen())
        stockfish_depth_15.set_fen_position(board.fen())

        # 推進棋局
        try:
            board.push_san(move)
        except Exception as e:
            print(f"Error at game {idx}, move: {move}")
            print(f"Error: {e}")
            break

        # 取得不同深度的評估
        eval_info_5 = stockfish_depth_5.get_evaluation()
        eval_info_15 = stockfish_depth_15.get_evaluation()

        # 儲存 value（只取 'cp' 或 'mate' 數值）
        if eval_info_5["type"] == "cp":
            evaluations_depth_5.append(eval_info_5["value"])
        elif eval_info_5["type"] == "mate":
            evaluations_depth_5.append(10000 if eval_info_5["value"] > 0 else -10000)
        else:
            evaluations_depth_5.append(None)

        if eval_info_15["type"] == "cp":
            evaluations_depth_15.append(eval_info_15["value"])
        elif eval_info_15["type"] == "mate":
            evaluations_depth_15.append(10000 if eval_info_15["value"] > 0 else -10000)
        else:
            evaluations_depth_15.append(None)

    # 儲存到 DataFrame
    df.at[idx, "Evaluation_Depth_5"] = evaluations_depth_5
    df.at[idx, "Evaluation_Depth_15"] = evaluations_depth_15

    # 每盤棋處理完，印棋局編號
    print(f"Game {idx} done")

# 結束計時
end_time = time.time()
total_time = end_time - start_time

print(f"\n全部處理完成！總花費時間：{total_time:.2f} 秒")

# 儲存結果
output_folder = "database/Filter_evaluate_CSV/180+2"  # ⚡記得換
os.makedirs(output_folder, exist_ok=True)
output_file_path = os.path.join(output_folder, "chunk_1_evaluated.csv")
df.to_csv(output_file_path, index=False)
print(f"儲存到 {output_file_path} 完成！")
```

```python
def analyze_game(moves):
    move_list = moves.split()  # Split PGN moves into a list
    stockfish.set_position([])  # Start from the initial position
    evaluations = []

    for move in move_list:
        if '.' in move:  # Skip move numbers (e.g., '1.' or '2.')
            continue
        stockfish.make_moves_from_current_position([move])  # Play the move
        eval_info = stockfish.get_evaluation()  # Get evaluation
        
        # Convert centipawns (cp) to pawn units (e.g., +1.0 instead of 100)
        if eval_info["type"] == "cp":
            score = eval_info["value"] / 100.0  # Convert cp to pawn units
        else:
            score = "#" + str(eval_info["value"])  # Mate in x moves

        evaluations.append((move, score))  # Store move + converted evaluation

    return evaluations  # Return list of (move, score)
```
